```c++
Double_t fPtDistr(Double_t* x,Double_t* par)
{
  /* - Par 0, 1, 2:   coherent.
     - Par 3, 4, 5:   incoherent.
     - Par 6      :   gamma+gamma.
     -
   */
  Double_t val = 0;
  val += par[0]* ( fIncohJpsiToMu     ->Interpolate(x[0]) );   //needed
  val += par[1]* ( fTwoGammaToMuMedium->Interpolate(x[0]) );   //needed
  val += par[2]* ( fHighPtTail        ->Interpolate(x[0]) );   //needed

  return val;
}
```
